```python
import pandas as pd

df = pd.read_csv("../data/WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

```python
df.shape
list(df.columns)
print(df.isnull().sum())
```

```python
df.info()
```

```python
df["Attrition"].value_counts()
```

```python
import matplotlib.pyplot as plt

df["Attrition"].value_counts().plot(kind="bar")
plt.title("Employee Attrition Count")
plt.show()
```

```python
df.groupby("Attrition")["MonthlyIncome"].mean()
df.groupby("Attrition")["Age"].mean()
df["OverTime"].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x="Attrition", y="MonthlyIncome", data=df)
plt.title("Salary vs Attrition")
plt.show()
```

```python
sns.barplot(x="Attrition", y="Age", data=df)
plt.title("Age vs Attrition")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
sns.countplot(data=df, x="JobRole", hue="Attrition")
plt.xticks(rotation=45)
plt.title("Job Role vs Attrition")
plt.show()
```

```python
sns.countplot(data=df, x="OverTime", hue="Attrition")
plt.title("OverTime vs Attrition")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(data=df, x="EducationField", hue="Attrition")
plt.xticks(rotation=45)
plt.title("Education Field vs Attrition")
plt.show()
```

```python
sns.boxplot(x="Attrition", y="MonthlyIncome", data=df)
plt.title("Salary Distribution vs Attrition")
plt.show()
```

```python
sns.countplot(data=df, x="JobSatisfaction", hue="Attrition")
plt.title("Job Satisfaction vs Attrition")
plt.show()
```

# Employee Attrition Analysis Project

## Objective
To analyze employee attrition and understand why employees leave the company.

## Tools Used
- Python
- Pandas
- Matplotlib
- Seaborn
- Jupyter Notebook

## Key Insights
- Low salary increases chances of leaving
- Overtime workers leave more
- Job satisfaction affects employee retention
- Certain job roles have higher attrition

## Conclusion
This project helps HR teams understand employee behavior and improve retention strategies.
